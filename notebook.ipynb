```python
gps = pd.read_csv("15_10.csv")
gps.columns=['segment', 'latitude', 'longitude', 'altitude', 
            'Time (UTC ISO)', 'time(UTC)', 'time(local)', 'UNIX Epoch Time']
print(gps.head())
```

```python
plt.rc('font', size=10)
plt.rcParams['figure.constrained_layout.use'] = True

plt.scatter(gps['longitude'], gps['latitude'], color='blue', marker='.')
plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
plt.rc('font', size=10)

fig = plt.figure(figsize = (7, 7))

ax = fig.add_subplot(111, projection='3d')
ax.scatter(gps['longitude'], gps['latitude'], gps['time(local)'], color='blue', marker='.')
ax.set_xlabel('longitude', fontweight ='bold')
ax.set_ylabel('latitude', fontweight ='bold')
ax.set_zlabel('time', fontweight ='bold')
ax.view_init(20, 75)
ax.dist = 10
```
